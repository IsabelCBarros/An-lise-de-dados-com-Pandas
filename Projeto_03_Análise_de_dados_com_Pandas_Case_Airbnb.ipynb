**DADOS ESTRUTURADOS**

Têm estruturas rígidas, ou seja, antes mesmo da presença dos dados ali, aquele ambiente já foi pensado para eles. Com isso, os estruturados não aceitam dados com estruturas diferentes feitas previamente.

Se criar uma tabela no Excel para ser exclusivamente textual, ela não aceitará números.

Tipos de dados estruturados
Existem diversos tipos de deles na internet, alguns deles são:

* Planilhas eletrônicas como as do Excel;
* Os próprios bancos de dados;
* Arquivos CSV;
* Arquivos XML;
* Arquivos JSON.

Têm um esquema fixo e consistente, o que facilita a consulta e a análise usando linguagens de consulta, como SQL (Structured Query Language).

**DADOS SEMIESTRUTURADOS**

Os dados semiestruturados não têm um esquema definido. Não cabem num formato de tabelas / linhas / colunas, mas são organizados por meio de etiquetas ou “tags” que permitem agrupá-los e criar hierarquias. São também conhecidos como não-relacionais ou NoSQL.
Este tipo de dados representa cerca de 5–10% do volume global de dados. No entanto, tem casos de utilização muito relevantes do ponto de vista comercial, em grandes infraestruturas de dados e aplicações web em tempo real.

Alguns serviços bem conhecidos baseados neste tipo de dados são o sistema de recomendação da Amazon, ou os serviços do Linkedin. Muitos dos casos de utilização envolvem transporte de dados, partilha de dados de sensores, intercâmbio eletrónico de dados, plataformas de redes sociais, e bases de dados NoSQL.

Exemplo de dados semi-estruturados:

* Emails, onde os metadados nativos permitem a classificação e pesquisa de  palavras-chave na linguagem de marcação
XML, cuja estrutura flexível, baseada em etiquetas, permite a estrutura universal de dados, armazenamento e transporte na Web.
* O padrão aberto JSON (JavaScript Object Notation), outro formato semiestruturado de intercâmbio de dados que é amplamente utilizado na transmissão de dados entre aplicações web e servidores.
* As bases de dados NoSQL, que não separam o esquema dos dados em si, são mais flexíveis. Permitem armazenar informação que não se enquadra bem no formato registo / tabela, por exemplo, texto de comprimento variável. Também facilitam o intercâmbio de dados entre diferentes bases de dados.

**ETL (Extrair, Transformar, Carregar)**

ETL é a sigla para o processo de extrair, transformar e carregar. É uma forma tradicionalmente aceita para que as organizações combinem dados de vários sistemas em um único banco de dados, repositório de dados, armazenamento de dados ou data lake. O ETL pode ser usado para armazenar dados legados, ou, o que é mais comum, agregar dados para analisar e impulsionar as decisões de negócios.

Fonte: https://cloud.google.com/learn/what-is-etl?hl=pt-br

Como funciona o processo?
*  Extração (E): nesta fase, os dados são coletados de diferentes sistemas
organizacionais e conduzidos a um espaço temporário (staging area - durante a execução da ETL em andamento), onde são convertidos em um mesmo formato para transformação.
*  Transformação (T): os dados brutos são lapidados e padronizados conforme as necessidades da empresa. Ao fim desta etapa, os dados estão “limpos”, estruturados e prontos para armazenamento.
*  Carregamento (L): os dados tratados são enviados a um repositório específico, onde serão armazenados em segurança e acionados para consulta interna.

**EDA - Exploratory Data Analysis (Análise Exploratória de Dados)**

É uma abordagem inicial na análise de dados, onde o principal objetivo é examinar e entender os dados antes de aplicar métodos estatísticos mais avançados ou construir modelos preditivos.

No processo de cada etapa da análise exploratório é importante fazer:

* Descrição de dados: identificação dos tipos de variáveis, dimensões dos dados, valores extremos e estatísticas descritivas.

* Visualização dos dados: criação de gráficos, tabelas e outros tipos de visualização para entender a distribuição e a relação entre as variáveis.

* Identificação de padrões: identificação de padrões, tendências, correlações e outras relações entre as variáveis.

* Testes estatísticos: aplicação de testes estatísticos para confirmar ou refutar hipóteses sobre os dados.

Com base nessas conclusões, algumas recomendações para o Airbnb melhorar a experiência dos anfitriões, hóspedes e consequentemente aumentar a rentabilidade do negócio, podem incluir:

* Investimento em maiores números de acomodações compartilhadas com um maior desconto, observando as medidas sanitárias e a segurança no momento de selecionar e aprovar pessoas desconhecidas para compartilharem da mesma acomodação;

* Pensar na possibilidade de oferecer acomodações compartilhadas a grupos específicos, tais como: aluguel de acomodações somente para homens, somente para mulheres, somente para o grupo LGBTQIA+, desde que as regras estejam claras e os hóspedes e anfitriões leiam e concordem com termos disponibilizados no site durante o procedimento de locação. Mapear esses perfis de interesse quando os hóspedes fizerem cadastro no site e ofertá-los regularmente, de acordo as suas preferências;

* Expandir a capacidade dos quartos privados para acomodar mais pessoas e oferececer descontos na locação, observando as medidas sanitárias e de acordo com as preferências dos hóspedes.

* Estabelecer um protocolo, promover e monitorar a limpeza dos imóveis, uma vez que uma boa avaliação nesse critério é valorizada pelos hóspedes.

* Melhorar a divulgação e oferta de imóveis com bons índices locais de atração e restaurante.

* Fazer parcerias com atrações, restaurantes, empresas de transportes, e oferecer cupons de descontos/fidelidade para os hóspedes.

* Incentivar os anfitriões a alcançar o status de Superhost com algumas ações, sendo elas:

>> *Educação e recursos: O Airbnb pode fornecer recursos educacionais e orientações claras sobre os critérios e expectativas para se tornar um Superhost. Isso pode incluir workshops, webinars, tutoriais online e diretrizes detalhadas sobre como atender aos requisitos necessários.*



>> *Reconhecimento e incentivos: Além dos incentivos oferecidos atualmente, poderia incluir descontos em taxas de serviço ou acesso antecipado a recursos e atualizações da plataforma.*

>> *Feedback e avaliações: O Airbnb pode revisar e ajustar seu sistema de avaliação e feedback para garantir que seja mais equilibrado e justo. Isso pode incluir a implementação de medidas para filtrar avaliações injustas ou imprecisas e permitir que os Superhosts tenham mais oportunidades de responder e resolver problemas antes de receber avaliações negativas.*

>> *Suporte ao anfitrião: Oferecer um suporte aprimorado aos anfitriões, com tempos de resposta rápidos e eficientes, canais de comunicação dedicados e resolução efetiva de problemas, pode ajudar a criar uma experiência positiva para os anfitriões e incentivar mais pessoas a se tornarem Superhosts.*

>> *Programas de incentivo e metas: O Airbnb pode criar programas de incentivo, como bônus financeiros, recompensas ou benefícios exclusivos para anfitriões que atinjam determinados critérios de desempenho e se tornem Superhosts. Isso pode motivar os anfitriões a se esforçarem mais para alcançar e manter esse status.*
