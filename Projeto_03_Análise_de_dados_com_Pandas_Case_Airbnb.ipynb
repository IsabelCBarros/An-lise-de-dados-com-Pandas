**ETL (Extrair, Transformar, Carregar)**

ETL  a sigla para o processo de extrair, transformar e carregar.  uma forma tradicionalmente aceita para que as organizaes combinem dados de vrios sistemas em um nico banco de dados, repositrio de dados, armazenamento de dados ou data lake. O ETL pode ser usado para armazenar dados legados, ou, o que  mais comum, agregar dados para analisar e impulsionar as decises de negcios.

Fonte: https://cloud.google.com/learn/what-is-etl?hl=pt-br

Como funciona o processo?
*  Extrao (E): nesta fase, os dados so coletados de diferentes sistemas
organizacionais e conduzidos a um espao temporrio (staging area - durante a execuo da ETL em andamento), onde so convertidos em um mesmo formato para transformao.
*  Transformao (T): os dados brutos so lapidados e padronizados conforme as necessidades da empresa. Ao fim desta etapa, os dados esto limpos, estruturados e prontos para armazenamento.
*  Carregamento (L): os dados tratados so enviados a um repositrio especfico, onde sero armazenados em segurana e acionados para consulta interna.
